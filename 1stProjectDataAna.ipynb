```python
import pandas as pd
import psycopg2
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
""" it by default gives the summary statistical values of only columns which has integer values :- but  if we use
include='all' keyword then it will also give the summary statistics of categorical values """
df.describe(include='all')
```

```python

df.isnull().sum()
```

```python
"""so now we have to handle the missing value by fullfilling null values for that we have use median method in code ."""
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#creating a column :-
labels = ['Adult','Young Adult','middle_aged','senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#create a new column purchase_frequency_days[it shows how frequently / oftenly customer shops.]
frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-weekly':14,
    'Annually':365,
    'Every 3 Months':90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```
